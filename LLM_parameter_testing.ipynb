```python
import torch
print("CUDA available:", torch.cuda.is_available())
print("Device:", torch.device("cuda" if torch.cuda.is_available() else "cpu"))
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen3-0.6B")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen3-0.6B")
```

```python
messages = [
    {"role": "user", "content": "write story on machine learning"},
]
inputs = tokenizer.apply_chat_template(
	messages,
	add_generation_prompt=True,
	tokenize=True,
	return_dict=True,
	return_tensors="pt",
).to(model.device)
```

```python
with torch.no_grad():
  output = model.generate(**inputs, max_new_tokens=100,temperature=0.1, top_p=0.95,repetition_penalty=1.2, do_sample=True,top_k=5)
print(tokenizer.decode(output[0][inputs["input_ids"].shape[-1]:]))
```
