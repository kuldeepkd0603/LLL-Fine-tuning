```python
import os
import json
import torch
import wandb
import gc
from google.colab import drive
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig,
    TrainingArguments
)
from peft import LoraConfig, get_peft_model
from trl import SFTTrainer, SFTConfig
import warnings
warnings.filterwarnings("ignore")
```

```python
drive.mount('/content/drive')

wandb.login()

wandb.init(
    project="tinyllama-chat-finetune",
    name="tinyllama-lora-experiment",
    config={
        "model_name": "TinyLlama/TinyLlama-1.1B-Chat-v1.0",
        "lora_r": 8,
        "lora_alpha": 16,
        "max_seq_length": 512,
        "batch_size": 1,
        "learning_rate": 5e-5,
        "num_train_epochs": 3
    }
)

os.environ["WANDB_LOG_MODEL"] = "checkpoint"
```

```python
def format_chat_messages(messages_list):
    """
    Convert messages to TinyLlama chat format
    """
    formatted_data = []

    for conversation in messages_list:
        # Convert to the chat format expected by TinyLlama
        text = ""
        for message in conversation["messages"]:
            if message["role"] == "system":
                text += f"<|system|>\n{message['content']}</s>\n"
            elif message["role"] == "user":
                text += f"<|user|>\n{message['content']}</s>\n"
            elif message["role"] == "assistant":
                text += f"<|assistant|>\n{message['content']}</s>\n"

        formatted_data.append({"text": text})

    return formatted_data

# Load your JSONL data
def load_chat_data(file_path):
    """
    Load chat data from JSONL file
    """
    data = []
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            data.append(json.loads(line))
    return data

data_path = "/content/chat_ready_data.jsonl"
chat_data = load_chat_data(data_path)
formatted_data = format_chat_messages(chat_data)

dataset = Dataset.from_list(formatted_data)
print(f"Dataset size: {len(dataset)}")
print(f"Sample: {dataset[0]['text'][:200]}...")
```

```python
# Model configuration for memory efficiency
model_name = "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

# BitsAndBytes configuration for 4-bit quantization
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
tokenizer.padding_side = "right"

# Load model with quantization
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    torch_dtype=torch.bfloat16,
    trust_remote_code=True
)

print(f"Model loaded: {model_name}")
print(f"Model device: {next(model.parameters()).device}")
```

```python
# LoRA configuration optimized for TinyLlama
lora_config = LoraConfig(
    r=8,                        # Low rank
    lora_alpha=16,              # LoRA scaling parameter
    target_modules=[            # Target modules for TinyLlama
        "q_proj",
        "v_proj",
        "k_proj",
        "o_proj",
        "gate_proj",
        "up_proj",
        "down_proj",
    ],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

# Apply LoRA to model
model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

# Enable gradient checkpointing for memory efficiency
model.gradient_checkpointing_enable()
```

```python
training_args = SFTConfig(
    output_dir="/content/drive/MyDrive/tinyllama-lora-checkpoint",
    per_device_train_batch_size=1,
    gradient_accumulation_steps=4,
    warmup_steps=10,
    max_steps=100,
    learning_rate=5e-5,
    fp16=False,
    bf16=True,
    logging_steps=5,
    optim="adamw_torch",
    weight_decay=0.001,
    lr_scheduler_type="linear",
    seed=42,
    max_seq_length=512,
    report_to="wandb",
    save_steps=25,
    save_total_limit=2,
    dataloader_drop_last=True,
    gradient_checkpointing=True,
)

gc.collect()
torch.cuda.empty_cache()
```

```python
trainer = SFTTrainer(
    model=model,
    processing_class=tokenizer,
    train_dataset=dataset,
    peft_config=lora_config,
    # dataset_text_field="text",
    args=training_args,
    # max_seq_length=512,
)

model.train()
model = get_peft_model(model, lora_config)
model.enable_input_require_grads()
model.train()
print("Trainable parameters:")
for n, p in model.named_parameters():
    if p.requires_grad:
        print(n)
assert any(p.requires_grad for p in model.parameters()), "No trainable parameters!"


print("Is train mode?", model.training)
print("Starting training...")


trainer.train()

gc.collect()
torch.cuda.empty_cache()
```

```python
def test_model(model, tokenizer, prompt, max_new_tokens=100):
    """
    Test the fine-tuned model
    """
    model.eval()
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_new_tokens=max_new_tokens,
            temperature=0.7,
            do_sample=True,
            top_p=0.9,
            pad_token_id=tokenizer.eos_token_id
        )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response


test_prompt = "<|user|>\nwhich types of movies .</s>\n<|assistant|>\n"
response = test_model(model, tokenizer, test_prompt)
print(f"Model response: {response}")


wandb.log({
    "test_prompt": test_prompt,
    "test_response": response
})
```

```python
import os
import json
import torch
import wandb
import gc
from google.colab import drive
from datasets import Dataset
from transformers import (
    AutoTokenizer,
    AutoModelForCausalLM,
    BitsAndBytesConfig,
    TrainingArguments
)
from peft import LoraConfig, get_peft_model
from trl import SFTTrainer, SFTConfig
import warnings
warnings.filterwarnings("ignore")
```

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype="bfloat16"
)
```

```python
def load_finetuned_model(base_model_name, adapter_path):
    """
    Load the fine-tuned model for inference
    """
    from peft import PeftModel
    # Load base model
    base_model = AutoModelForCausalLM.from_pretrained(
        base_model_name,
        quantization_config=bnb_config,
        device_map="auto",
        torch_dtype=torch.bfloat16
    )

    # Load LoRA adapters
    model = PeftModel.from_pretrained(base_model, adapter_path)

    # Load tokenizer
    tokenizer = AutoTokenizer.from_pretrained(adapter_path)

    return model, tokenizer

model, tokenizer = load_finetuned_model(
    "TinyLlama/TinyLlama-1.1B-Chat-v1.0",
    "/content/drive/MyDrive/tinyllama-lora-finetuned"
)


input_text = "i ma fine"
input_ids = tokenizer(input_text, return_tensors="pt").input_ids.to(model.device)
output = model.generate(input_ids, max_new_tokens=100)
print(tokenizer.decode(output[0], skip_special_tokens=True))
```

```python
chat_history = []

def get_prompt(chat_history):
    prompt = ""
    for message in chat_history:
        if message["role"] == "user":
            prompt += f"<|user|>\n{message['content']}</s>\n"
        elif message["role"] == "assistant":
            prompt += f"<|assistant|>\n{message['content']}</s>\n"
    prompt += "<|assistant|>\n"
    return prompt

print("Welcome to the chat! Type 'exit' to quit.")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break
    chat_history.append({"role": "user", "content": user_input})

    prompt = get_prompt(chat_history)

    input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to(model.device)
    output = model.generate(input_ids, max_new_tokens=100)
    response = tokenizer.decode(output[0], skip_special_tokens=True)

    if response.startswith(prompt):
        response = response[len(prompt):]
    assistant_reply = response.strip()
    print("Assistant:", assistant_reply)
    print("============================")

    chat_history.append({"role": "assistant", "content": assistant_reply})
```
